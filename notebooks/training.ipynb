```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df_scaled = pd.read_csv('../data/valorant_scaled_data.csv')
X_train = df_scaled.drop(columns=['player_id', 'archetype'])
```

```python
# Train the K-Means model on the unlabeled features
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
discovered_clusters = kmeans.fit_predict(X_train)
```

```python
# Load the original unscaled data so we can interpret the raw numbers of the new clusters
df_original = pd.read_csv('../data/valorant_synthetic_data_raw.csv')
```

```python
# Assign the model's discovered cluster numbers back to the original unscaled dataframe
df_original['assigned_cluster'] = discovered_clusters
```

```python
# Calculate the average stats for each discovered cluster to understand the playstyles it found
cluster_profiles = df_original.groupby('assigned_cluster').mean(numeric_only=True)
display(cluster_profiles)
```

### Interpreting the Discovered Clusters
In a real-world scenario, you look at the output of Cell 6 to name the clusters yourself. 

For example, if the output shows that `Cluster 0` has an average `first_bloods_per_round` of 0.18 and a `sniper_kill_percent` of 25%, you would map `Cluster 0` to the string "Aggressive Entry" in your backend. If `Cluster 2` has the highest `assists_per_round` and lowest `adr`, you map it to "Utility-Heavy".

This proves the model can successfully take blind data and organize players into distinct, recognizable Valorant playstyles.
