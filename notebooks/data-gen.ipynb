```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_players = 1000 
```

```python
def generate_archetype(name, adr_mean, kast_mean, fb_mean, fd_mean, apr_mean, clutch_mean, surv_mean, trade_mean, hs_mean, econ_mean, sniper_mean):
    return {
        "archetype": name,
        "adr": np.clip(np.random.normal(adr_mean, 15, n_players), 50, 300),
        "kast_percent": np.clip(np.random.normal(kast_mean, 5, n_players), 30, 95),
        "first_bloods_per_round": np.clip(np.random.normal(fb_mean, 0.04, n_players), 0, 0.4),
        "first_deaths_per_round": np.clip(np.random.normal(fd_mean, 0.04, n_players), 0, 0.4),
        "assists_per_round": np.clip(np.random.normal(apr_mean, 0.1, n_players), 0, 1.2),
        "clutch_win_rate": np.clip(np.random.normal(clutch_mean, 5, n_players), 0, 100),
        "survival_rate": np.clip(np.random.normal(surv_mean, 5, n_players), 0, 100),
        "trade_percent": np.clip(np.random.normal(trade_mean, 3, n_players), 0, 40),
        "headshot_percent": np.clip(np.random.normal(hs_mean, 4, n_players), 5, 50),
        "econ_rating": np.clip(np.random.normal(econ_mean, 10, n_players), 20, 120),
        "sniper_kill_percent": np.clip(np.random.normal(sniper_mean, 5, n_players), 0, 80)
    }
```

```python
entry_data = generate_archetype("Aggressive Entry", 165, 70, 0.18, 0.18, 0.15, 12, 18, 10, 22, 55, 25)
support_data = generate_archetype("Passive Support", 125, 78, 0.06, 0.06, 0.45, 15, 35, 22, 20, 65, 5)
lurker_data = generate_archetype("Lurker", 150, 72, 0.08, 0.05, 0.20, 25, 30, 8, 28, 75, 8)
utility_data = generate_archetype("Utility-Heavy", 110, 75, 0.05, 0.08, 0.55, 10, 28, 15, 16, 60, 2)
flex_data = generate_archetype("Balanced Flex", 135, 72, 0.11, 0.11, 0.30, 15, 25, 14, 21, 62, 10)
```

```python
# shuffle and save
datasets = [entry_data, support_data, lurker_data, utility_data, flex_data]
df_list = [pd.DataFrame(data) for data in datasets]
df = pd.concat(df_list, ignore_index=True)
df.insert(0, 'player_id', ['player_' + str(i) for i in range(1, len(df) + 1)])
df = df.sample(frac=1).reset_index(drop=True)

df.to_csv('../data/valorant_synthetic_data_raw.csv', index=False)
print("dataset created successfully")
```
