```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('../data/valorant_synthetic_data_raw.csv')
df_sample = df.sample(frac=0.2, random_state=42).reset_index(drop=True)
```

```python
features = df.drop(columns=['player_id', 'archetype'])
metadata = df[['player_id', 'archetype']]
```

```python
# Scale the features so all metrics have a mean of 0 and variance of 1
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)
df_final = pd.concat([metadata, df_scaled], axis=1)
df_final.to_csv('../data/valorant_scaled_data.csv', index=False)
```

```python
df_sample = df_final.sample(frac=0.2, random_state=42).reset_index(drop=True)
```

```python
# Extract just the scaled features for the unsupervised models
X_sample = df_sample.drop(columns=['player_id', 'archetype'])
```

```python
# Train and evaluate K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_sample)
print(f"K-Means Silhouette Score: {silhouette_score(X_sample, kmeans_labels):.4f}")
```

```python
# Train and evaluate Gaussian Mixture Model
gmm = GaussianMixture(n_components=5, random_state=42)
gmm_labels = gmm.fit_predict(X_sample)
print(f"GMM Silhouette Score: {silhouette_score(X_sample, gmm_labels):.4f}")
```
